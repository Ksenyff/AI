```python
import numpy as np

# пустой двумерный массив
sample = np.empty((0,2))

# заполнение массива именами мальчиков и соответствующей классификацией 0-м
with open("/content/drive/MyDrive/Classroom/AI Intro/Второе домашнее задание (спам-фильтр)/data/boys+.txt", "r") as file1:
    data = file1.readlines()
    for line in data:
        words = line.split(',')
        sample = np.append(sample, [[words[0], words[1][0]]], axis=0)

# заполнение массива именами девочек и соответствующей классификацией 1-ж
with open("/content/drive/MyDrive/Classroom/AI Intro/Второе домашнее задание (спам-фильтр)/data/girls+.txt", "r") as file1:
    data = file1.readlines()
    for line in data:
        words = line.split(',')
        sample = np.append(sample, [[words[0], words[1][0]]], axis=0)


# перемешивание элементов в массиве
dataset = np.random.permutation(sample)

# создание массива фич, где фичей выступает последняя буква имени, а лэйблом классификация (м и ж)
features = [(feat[-1], int(label)) for feat, label in dataset]
print(features)
```

```python
# метод расчитывающий классификатор (соотношение лэйблов в массиве и вероятность принадлежности к некоторому лэйблу при наличии некоторой фичи)
def fit(dataset):
    classes, freq = {}, {}
    for feats, label in dataset:
        # подсчет количества представителей лейбла
        if label not in classes:
            classes[label]=0
        classes[label]+=1
        # подсчет количества представителей фичи в некотором лэйбле
        for feat in feats:
            if (label, feat) not in freq:
                freq[(label, feat)] = 0
            freq[(label, feat)] +=1
    # расчет соотношения количества представителей фичи в лэйбле к представителям лэйбла
    for label, feat in freq:
            freq[(label, feat)] /= classes[label]
    # расчет соотношения количества представителей лэйбла к общему количеству объектов
    for c in classes:
            classes[c] /= len(dataset)

    return classes, freq

fit(features)
```

```python
classifier = fit(features)

import math


# метод для определения принадлежности некоторому лэйблу по заданой фиче и имеющемся классификаторе
def classify(classifier, features):
    classes, freq = classifier
    return min(classes.keys(), key = lambda cls: -math.log(classes[cls]) + sum(-math.log(freq.get((cls, feat), 10**(-7))) for feat in features))

classify(classifier, "а")
```

```python
# метод для выделения новой фичи (первая буква и последние две буквы имени)
def get_features(sample):
    features = sample[0] + sample[-2] + sample[-1]
    return features.lower()

# выделение фич в датасете
features = [([get_features(feat)], int(label)) for feat, label in dataset]
# создание классификатора
classifier = fit(features)
# предсказание принадлежности лэйблу(классу) некоторого имени(с предварительно выделеной фичей)
print(classify(classifier, [get_features("Сергей")]))
```

```python
# массив сообщений спама
spam = ["Путевки по низкой цене", "Акция! Купи шоколадку и получи телефон в подарок"]
# массив сообщений с важной информацией
imp = ["Завтра состоится собрание", "Купи килограмм яблок и шоколадку"]
# массив с тестовой выборкой
test = ["В магазине гора яблок. Купи семь килограмм и шоколадку"]

# массив с шумом (знаками и словами с равной вероятностью встречающимися во всех видах сообщений)
noise = ['!', '?', '.', ',', ':']

# метод для превращения предложений в список отдельных слов с удалением шума
def make_dict(list1):
    result = []
    for sent in list1:
        words = sent.split()
        for word in words:
            for n in noise:
                word = word.replace(n,'')
            if len(word) >3:
                result += [word.lower()]
    return result

# заполнение датасета наборами слов спама и важных сообщений с соответствующими классификациями (0 - спам, 1 - важное)
dataset = []
for item in make_dict(spam):
    dataset.append(([item],0))

for item in make_dict(imp):
    dataset.append(([item],1))

# метод для создания классификатора на основании наивного байессовского метода
def fit(dataset, alpha):
    classes, freq, tot = {}, {}, set()
    for feats, label in dataset:
        if label not in classes:
            classes[label]=0
        classes[label]+=1
        for feat in feats:
            if (label, feat) not in freq:
                freq[(label, feat)] = 0
            freq[(label, feat)] +=1
        tot.add(tuple(feats))

    for label, feat in freq:
            freq[(label, feat)] =(alpha + freq[(label, feat)]) /( alpha * len(tot) + classes[label])
    for c in classes:
            classes[c] /= len(dataset)

    return classes, freq

# загрузка датасета для создания классификатора
classifier = fit(dataset, 0)

# классификация тестовой выборки
print(classify(classifier, make_dict(test)))
print(classify(classifier, make_dict(["Купи яблок по акции и получи телефон"])))
```
